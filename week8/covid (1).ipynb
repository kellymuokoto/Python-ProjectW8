```python
df = df_all[["date", "location", "total_cases", "total_deaths", "new_cases", "new_deaths", "total_vaccinations", "population"]]
```

```python
df.head()
```

```python
df_all.location.unique()
```

```python
df = df[df["location"].isin(["South Africa", "Zimbabwe", "United States", "India", "Kenya", "Nigeria", "China", "United Kingdom"])]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.drop_duplicates(inplace=True)
```

```python
df = df.dropna(subset=["date", "location", "total_cases", "total_deaths", "new_cases", "new_deaths", "total_vaccinations"])
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.fillna(0, inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="date", y="total_cases", hue="location")
plt.title("Total Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="date", y="total_deaths", hue="location")
plt.title("Total Deaths Over Time")
plt.xlabel("Date")
plt.ylabel("Total Deaths")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="date", y="new_cases", hue="location")
plt.title("Daily New Cases Comparison")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.show()
```

```python
df["death_rate"] = df["total_deaths"] / df["total_cases"]

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="date", y="death_rate", hue="location")
plt.title("COVID-19 Death Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Death Rate")
plt.show()
```

```python
latest_df = df[df["date"] == df["date"].max()]
top_cases = latest_df.groupby("location")["total_cases"].max().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_cases.index, y=top_cases.values)
plt.title("Top Countries by Total Cases (Latest Date)")
plt.xlabel("Country")
plt.ylabel("Total Cases")
plt.xticks(rotation=45)
plt.show()
```

```python
numeric_cols = ["total_cases", "total_deaths", "new_cases", "new_deaths", "death_rate"]
correlation = df[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="date", y="total_vaccinations", hue="location")
plt.title("Cumulative Vaccinations Over Time")
plt.xlabel("Date")
plt.ylabel("Total Vaccinations")
plt.show()
```

```python
vaccinated = df["population"].max()
```

```python

df["percent_vaccinated"] = (df["total_vaccinations"] / df["population"]) * 100
```

```python
latest = df[df["date"] == df["date"].max()]
percent_data = latest.groupby("location")["percent_vaccinated"].max()

plt.figure(figsize=(10, 6))
sns.barplot(x=percent_data.index, y=percent_data.values)
plt.title("Percent of Population Vaccinated")
plt.xlabel("Country")
plt.ylabel("% Vaccinated")
plt.xticks(rotation=45)
plt.show()
```

```python
import plotly.express as px


latest_date = df["date"].max()
latest_df = df[df["date"] == latest_date]


map_df = df[[ "location", "total_cases", "total_vaccinations", "population"]].copy()


map_df["cases_per_million"] = (map_df["total_cases"] / map_df["population"]) * 1_000_000
map_df["vaccination_rate"] = (map_df["total_vaccinations"] / map_df["population"]) * 100
```

```python
fig = px.choropleth(
    map_df,
    locations="location",
    locationmode="country names",
    color="cases_per_million",
    hover_name="location",
    color_continuous_scale="Reds",
    title="COVID-19 Cases per Million People (Latest Date)"
)
fig.show()
```

```python
fig = px.choropleth(
    map_df,
    locations="location",
    locationmode="country names",
    color="vaccination_rate",
    hover_name="location",
    color_continuous_scale="Greens",
    title="COVID-19 Vaccination Rate by Country"
)
fig.show()
```
